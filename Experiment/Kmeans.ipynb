```
sentences = nltk.sent_tokenize(data)
```

```
def word_embedding(sentences, model, unknown_emb):
    embeded = 0
    word_list = ViTokenizer.tokenize(sentences).split()
    for i in range(len(word_list)):
        if ((word_list[i] in model.index2word) == True):
            embeded = embeded + model.get_vector(word_list[i])
        else:
            embeded = embeded + unknown_emb
    return embeded
```

```
def doc_embedding(sentences):
    return model.infer_vector(nltk.word_tokenize(sentences))
```

```
# Load bao moi word2vec pretrained
word2vec_model = KeyedVectors.load_word2vec_format('./data/baomoi.window2.vn.model.bin', binary=True)
embedd_vectors = word2vec_model.vectors
unknown_embedd = np.zeros(300)
```

```
# Load wiki word2vec pretrained
wiki_word2vec_model = KeyedVectors.load_word2vec_format('./data/wiki.vi.model.bin', binary=True)
wiki_embedd_vectors = word2vec_model.vectors
wiki_unknown_embedd = np.zeros(400)
```

```
#Create vector tfidf
tf=TfidfVectorizer(min_df=3,max_df=0.88,max_features=3000,sublinear_tf=True)
tf.fit(sentences)
tfidf_vec=tf.transform(sentences)
newtf = scipy.sparse.csr_matrix.toarray(tfidf_vec)
```

```
#Load doc2vec pretrained
model= Doc2Vec.load("./data/d2v.model")
```

```
index = []
def vec(data,mode):
    data = data.lower()
    sentences = nltk.sent_tokenize(data)
    vec = []
    if (mode == 'all'):
        for i in range(len(sentences)):
            concated = np.concatenate((word_embedding(sentences[i], word2vec_model, unknown_embedd),newtf[i],doc_embedding(sentences[i])),axis=0)
            vec.append(concated)
            index.append(i)
    if (mode == 'd2v'):
        for i in range(len(sentences)):
            vec.append(doc_embedding(sentences[i]))
            index.append(i)
    if (mode == 'w2v'):
        for i in range(len(sentences)):
            vec.append(word_embedding(sentences[i], word2vec_model, unknown_embedd))
            index.append(i) 
    if (mode == 't2v'):
        for i in range(len(sentences)):
            vec.append(newtf[i])
            index.append(i)
    if (mode =='wiki_w2v'):
        for i in range(len(sentences)):
            vec.append(word_embedding(sentences[i], wiki_word2vec_model, wiki_unknown_embedd))
    return vec
```

```
def summary(n,vector):
    avg = []
    n_clusters = len(sentences)//n
    kmeans = KMeans(n_clusters=n_clusters)
    kmeans = kmeans.fit(vector)
    for j in range(n_clusters):
        idx = np.where(kmeans.labels_ == j)[0]
        avg.append(np.mean(idx))
    closest, _ = pairwise_distances_argmin_min(kmeans.cluster_centers_, vector)
    ordering = sorted(range(n_clusters), key=lambda k: avg[k])
    summary = ' '.join([sentences[closest[idx]] for idx in ordering])
    return summary
```
